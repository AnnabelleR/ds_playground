```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_string = fake_news.title
```

```python
MY_STOPWORDS = STOPWORDS | {'nan', 'video', 'comment'}
```

```python
print(MY_STOPWORDS)
```

```python
my_string = ' '.join(str(x).lower() for x in word_string)
```

```python
my_word_list = my_string.split()
```

```python
import re 

my_dic = {}

for x in my_word_list:
    x = re.sub(r'[^\w\s]','',x)
    if (x not in MY_STOPWORDS and x != ''):
        if x in my_dic:
            my_dic[x] += 1
        else:
            my_dic[x] = 1
```

```python
#[(k,v) for v,k in sorted(
#   [(v,k) for k,v in my_dic.items()],reverse=True
#   )
#]
```

```python
wordcloud = WordCloud(
                          stopwords=MY_STOPWORDS,
                          background_color='white',
                          width=1200,
                          height=1000
                         ).generate(my_string)


plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
